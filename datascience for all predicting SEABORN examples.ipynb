```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# seaborn is a visualization library
# seaborn can be installed by pip install seaborn
# required dependies are numpy,scipy,matplotlib,statsmodels,pandas

#in seaborn palettte attribute for color change. like winter_r and spring

# black lines are error bars
```

```python
#loading inbuilt tips dataset in seaborn
tips = sns.load_dataset('tips')
sns.barplot(x='day',y='tip',data = tips)
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
sns.barplot(x='day',y='tip',data=tips,hue='sex',palette='winter_r')
```

```python
sns.barplot(x='day',y='tip',data=tips,hue='smoker',palette='winter_r')
```

```python
# for horizontal visualization
sns.barplot(x='total_bill',y='day',data = tips,palette='spring')
```

```python
sns.barplot(x='day',y='tip',data=tips,palette= 'spring')
```

```python
# 
from numpy import median
sns.barplot(x='day',y='tip',data = tips ,estimator=median,palette='spring')
```

```python
sns.barplot(x='smoker',y='tip',data = tips, estimator=median,hue='sex',palette='coolwarm')
```
