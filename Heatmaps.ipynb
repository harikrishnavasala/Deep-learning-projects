```python
import numpy as np
import pandas as pd
import matplotlib
%matplotlib inline
```

```python
# taking normal distribution
normal = np.random.rand(12,14)
```

```python
import seaborn as sns
```

```python
sns.heatmap(normal)
```

```python
# in above figure higher value corresponds to darker value
# if we wanna see randomly distributed numbers
```

```python
sns.heatmap(normal,annot = True)
```

```python
# so the range is from 12 to 14 therefor distributtion between 0 and 2 
# to see heatmap from 0 to 2 corresponding values
```

```python
sns.heatmap(normal,vmin = 0,vmax = 2)
```

```python
sns.heatmap(normal,vmin = 0,vmax = 1)
```

```python
sns.heatmap(normal,vmin = 0,vmax = 0.5)
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
# tails gives last number of samples
flights.tail()
```

```python
flights = flights.pivot('month','year','passengers')
```

```python
sns.heatmap(flights)
```

```python
sns.heatmap(flights,annot = True)
```

```python
# to see the values clearly
sns.heatmap(flights,annot = True , fmt = 'd')
```

```python
sns.heatmap(flights, linewidths = 0.9)
```

```python
sns.heatmap(flights,linewidth = 0.9 ,annot = True , fmt = 'd')
```

```python
sns.heatmap(flights, linewidths = 1 ,annot = True , fmt = 'd' , vmin = 100 , vmax = 700)
```

```python
sns.heatmap(flights, cmap = 'RdBu',annot = True ,fmt = 'd')
```

```python
sns.heatmap(flights, cmap = 'summer',annot = True ,fmt = 'd')
```

```python
sns.heatmap(flights, cmap = 'RdBu',center = flights.loc['June',1954],annot = True ,fmt = 'd')
```

```python
sns.heatmap(flights,cbar = False)
```

```python
sns.heatmap(flights,cbar = False , linewidths = 1)
```
