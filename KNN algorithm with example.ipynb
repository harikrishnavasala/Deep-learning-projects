```python
# values of columns like 'glucose' , ' blood pressure' cannot be accepted as zeroes and 0 values 
#           are replaced with mean of respective column
```

```python
# replace zeroes
zero_not_accepted = ['Glucose','BloodPressure','SkinThickness','BMI','Insulin']
for column in zero_not_accepted:
    dataset[column] = dataset[column].replace(0,np.NaN) # NaN means no data
    mean = int(dataset[column].mean(skipna=True))
    dataset[column] = dataset[column].replace(np.NaN,mean)
    
```

```python
print(dataset['SkinThickness'])
```

# now split the data into train and test

```python
#split dataset
X = dataset.iloc[:,0:8] # this indicates in pandas all rows and columns from 0 to 8 
y = dataset.iloc[:,8]   # all rows and only single 8 column
X_train , X_test , y_train , y_test = train_test_split(X,y,random_state = 0 ,test_size =0.2)
```

```python
X_train
```

# perform feature scaling

```python
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
X_train
```

```python
# see above values are between 0 and 1
```

```python
# define the model init: k-nn
```

```python
classifier = KNeighborsClassifier(n_neighbors = 100, p=2 , metric = 'euclidean')
```

```python
import math
math.sqrt(len(y_test)) # here number is 12 and 12-1 is the number of neighbors
```

```python
# now fit the model
classifier.fit(X_train,y_train)
```

```python
# predict the test results
y_pred = classifier.predict(X_test)
y_pred
```

```python
# evaluate model
cm = confusion_matrix(y_test , y_pred)
cm
```

```python
print(f1_score(y_test,y_pred))
```
